- 표준점수
  - 가장 널리 사용하는 전처리 방법

```python
mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)
```

```python
print(mean, std)
```

```python
# 브로드캐스팅, 모든 행에 표준점수 적용

train_scaled = (train_input - mean) / std
```

#### **전처리 데이터로 모델 훈련**

```python
# 샘플을 스케일링 하지 않아서 동떨어져 있음

plt.scatter(train_scaled[:, 0], train_scaled[:, 1]) # 스케일링 된 좌표 (평균 = 0, 표준편차 = 1)
plt.scatter(25, 150, marker='^') # 스케일링 안 된 좌표
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
new = ([25, 150] - mean) / std

plt.scatter(train_scaled[:, 0], train_scaled[:, 1])
plt.scatter(new[0], new[1], marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 스케일링 한 훈련 데이터로 fit

kn.fit(train_scaled, train_target)
```

```python
# 테스트 데이터도 스케일링

test_scaled = (test_input - mean) / std
```

```python
kn.score(test_scaled, test_target)
```

```python
print(kn.predict([new]))
```

```python
distances, indexes = kn.kneighbors([new])
plt.scatter(train_scaled[:, 0], train_scaled[:, 1])
plt.scatter(new[0], new[1], marker='^')
plt.scatter(train_scaled[indexes, 0], train_scaled[indexes, 1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

## **3.1 k-최근접 이웃 회귀**

- 회귀 문제 다루기

```python
import numpy as np

# 농어 데이터

perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])
```

```python
import matplotlib.pyplot as plt

plt.scatter(perch_length, perch_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state=42)
```

```python
# 2차원 배열로 변환 = ML 모델은 입력이 무조건 2차원 배열이어야 함 (특성을 가진 샘플, 특성)

train_input = train_input.reshape(-1, 1) # 행 개수 자동, 열 1개
test_input = test_input.reshape(-1, 1)

print(train_input.shape, test_input.shape)
```

- 결정계수
  - 회귀 평가 방식
  - $R^2=1-\frac{(타깃-예측)^2의 합}{ (타킷-평균)^2의 합}$

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor()
knr.fit(train_input, train_target)
```

```python
print(knr.score(test_input, test_target))
```

```python
from sklearn.metrics import mean_absolute_error

# 타깃 값과 예측 값 사이 절댓값 오차 평균을 구하여 예측이 벗어난 정도를 측정 (결정계수보다 직관적으로 보기 위해)

test_prediction = knr.predict(test_input)
mae = mean_absolute_error(test_target, test_prediction)
print(mae)
```

```python
# 과적합 여부 판정 위해 훈련 데이터의 점수 측정
# 현재 훈련 데이터보다 (0.96) 테스트 세트의 점수가 (0.99) 더 높으므로 과소 적합 (= 덜 학습된 상태)

print(knr.score(train_input, train_target))
```

```python
# K 값이 너무 커서 훈련 데이터를 충분히 학습하지 못했으므로 K 값을 5에서 3으로 낮춤

knr.n_neighbors = 3
```

```python
knr.fit(train_input, train_target)

print(knr.score(train_input, train_target))
```

```python
# 훈련 데이터의 점수가 더 높아짐, 과소적합 해결

print(knr.score(test_input, test_target))
```
