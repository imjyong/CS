### **k-최근접 이웃 회귀**

```python
# 정답 분리

input = df.drop("quality", axis=1)
target = df["quality"]
```

```python
df.head()
```

```python
input.head()
```

```python
print(input.shape)
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(input, target, random_state=42)
```

```python
import numpy as np

# 각 열의 평균, 표준편차

mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)
```

```python
train_scaled = (train_input - mean) / std
test_scaled = (test_input - mean) / std
```

```python
from sklearn.neighbors import KNeighborsRegressor

# K = 5

knr = KNeighborsRegressor(n_neighbors=5)
knr.fit(train_scaled, train_target)
```

```python
# 결정계수

print(knr.score(test_scaled, test_target))
```

```python
from sklearn.metrics import mean_absolute_error

# 평균 절댓값 오차

test_prediction = knr.predict(test_scaled)
mae = mean_absolute_error(test_target, test_prediction)

print(mae)
```

```python
# K = 4

knr = KNeighborsRegressor(n_neighbors=4)
knr.fit(train_scaled, train_target)
```

```python
test_prediction = knr.predict(test_scaled)
mae = mean_absolute_error(test_target, test_prediction)

print(mae)
```

```python
# K = 3

knr = KNeighborsRegressor(n_neighbors=3)
knr.fit(train_scaled, train_target)
```

```python
test_prediction = knr.predict(test_scaled)
mae = mean_absolute_error(test_target, test_prediction)

print(mae)
```

```python
# K = 2

knr = KNeighborsRegressor(n_neighbors=2)
knr.fit(train_scaled, train_target)
```

```python
test_prediction = knr.predict(test_scaled)
mae = mean_absolute_error(test_target, test_prediction)

print(mae)
```
